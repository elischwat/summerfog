```python
#learning how to read csv files into python

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
tempdatafile = 'tempdata.csv'
tempdata = pd.read_csv(tempdatafile)

tempdata.rename(columns={
    'Date Time, GMT-07:00': 'DateTime',
    'Temp, °F (LGR S/N: 10567340, SEN S/N: 10567340)': 'Temp1',
    'Temp, °F (LGR S/N: 10567318, SEN S/N: 10567318)': 'Temp2',
    'Temp, °F (LGR S/N: 10759163, SEN S/N: 10759163)': 'Temp3',
    'Temp, °F (LGR S/N: 10567315, SEN S/N: 10567315)': 'Temp4',
    'Temp, °F (LGR S/N: 10567317, SEN S/N: 10567317)': 'Temp5'
}, inplace=True)

```

```python
tempdata.head()
# returns the first few rows of data
```

```python
f = plt.figure(figsize=(12, 6))


print(tempdata.columns)

tempdata["DateTime"] = pd.to_datetime(tempdata["DateTime"])

#plt.plot(tempdata['DateTime'], tempdata['Temp1'])
temp_columns = ['Temp1', 'Temp2', 'Temp3', 'Temp4', 'Temp5']


tempdata.set_index('DateTime')[temp_columns].plot()
plt.xlabel("Date Time")
plt.ylabel("Temperature (°F)")
plt.title("Temperature Trends")
plt.legend()
plt.show()


```

```python
#relative humidity
rhdatafile = 'rhdata.csv'
rhdata = pd.read_csv('rhdata.csv')

print(rhdata.columns)
rhdata.head()
```

```python
rhdata.rename(columns={
    'Date Time, GMT-07:00': 'DateTime',
}, inplace=True)
rhdata.head()
```

```python
print(rhdata.columns)

rhdata['DateTime'] = pd.to_datetime(rhdata['DateTime'])

#plt.plot(tempdata['DateTime'], tempdata['RH317'])

rh_columns = ['RH317', 'RH315', 'RH340', 'RH318', 'RH163']


print(rhdata[rh_columns].dtypes)
rhdata.set_index('DateTime')[rh_columns].plot()
plt.xlabel("Date Time")
plt.ylabel("RH (°F)")
plt.title("RH")
plt.legend()
plt.show()
```

```python
start = pd.Timestamp('2025-06-19 06:00:00')
end = pd.Timestamp('2025-06-19 12:00:00')

# Filter the data
filtered_data = rhdata[(rhdata['DateTime'] >= start) & (rhdata['DateTime'] <= end)]

# Plot the filtered data
filtered_data.set_index('DateTime')[rh_columns].plot(figsize=(12, 6))
plt.xlabel("Date Time")
plt.ylabel("RH (%)")
plt.title("Relative Humidity: June 19, 6AM to 12PM")
plt.legend()
plt.show()

#a 2.68% difference between RH315 and RH318
```
