```python
data_dict = {}

# Only include keys with actual sensor data (skip MATLAB metadata)

data_dict['PL_T'] = ds2['Temp']
data_dict['PL_datestr'] = pd.to_datetime(ds2['DateTime'])

data_dict['FHL_T'] = ds_fhl['Temp']
data_dict['FHL_datestr'] = pd.to_datetime(ds_fhl['DateTime'])

data_dict['CP1_T'] = ds['CP1_T'].flatten()
data_dict['CP2_T'] = ds['CP2_T'].flatten()
data_dict['CP3_T'] = ds['CP3_T'].flatten()
data_dict['CPL_T'] = ds['CPL_T'].flatten()
data_dict['CP1_datestr'] = pd.to_datetime(ds['CP1_datestr'].flatten())
data_dict['CP2_datestr'] = pd.to_datetime(ds['CP2_datestr'].flatten())
data_dict['CP3_datestr'] = pd.to_datetime(ds['CP3_datestr'].flatten())
data_dict['CPL_datestr'] = pd.to_datetime(ds['CPL_datestr'].flatten())


for i in range(0, 15724):
    if data_dict['CP1_datestr'][i] != data_dict['CP2_datestr'][i] and data_dict['CP1_datestr'][i] != data_dict['CP3_datestr'][i]:
        print('different dates found')

for k in data_dict:
    print(f"{k}: shape={data_dict[k].shape}, dtype={data_dict[k].dtype}")
```

```python
plt.figure(figsize=(10, 6))

plt.plot(data_dict['CP1_datestr'], data_dict['CP1_T'], label='CP1_T')
plt.plot(data_dict['CP2_datestr'], data_dict['CP2_T'], label='CP2_T')
plt.plot(data_dict['CP3_datestr'], data_dict['CP3_T'], label='CP3_T')
plt.plot(data_dict['CPL_datestr'], data_dict['CPL_T'], label='CPL_T')
plt.plot(data_dict['PL_datestr'], data_dict['PL_T'], label='PL_T')

plt.title('Temperature Readings from Multiple Sensors')
plt.xlabel('Date Time')
plt.ylabel('Temperature (°C)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
start_date = pd.Timestamp('2022-08-11')
end_date = pd.Timestamp('2022-08-12')

plt.figure(figsize=(10, 6))

# CP1_T
mask1 = (data_dict['CP1_datestr'] >= start_date) & (data_dict['CP1_datestr'] <= end_date)
plt.plot(data_dict['CP1_datestr'][mask1], data_dict['CP1_T'][mask1], label='CP1_T')

# CP2_T
mask2 = (data_dict['CP2_datestr'] >= start_date) & (data_dict['CP2_datestr'] <= end_date)
plt.plot(data_dict['CP2_datestr'][mask2], data_dict['CP2_T'][mask2], label='CP2_T')

# CP3_T
mask3 = (data_dict['CP3_datestr'] >= start_date) & (data_dict['CP3_datestr'] <= end_date)
plt.plot(data_dict['CP3_datestr'][mask3], data_dict['CP3_T'][mask3], label='CP3_T')

# CPL_T
mask4 = (data_dict['CPL_datestr'] >= start_date) & (data_dict['CPL_datestr'] <= end_date)
plt.plot(data_dict['CPL_datestr'][mask4], data_dict['CPL_T'][mask4], label='CPL_T')


# PL_T
mask5 = (data_dict['PL_datestr'] >= start_date) & (data_dict['PL_datestr'] <= end_date)
plt.plot(data_dict['PL_datestr'][mask5], data_dict['PL_T'][mask5], label='PL_T')


# FHL_T
mask6 = (data_dict['FHL_datestr'] >= start_date) & (data_dict['FHL_datestr'] <= end_date)
plt.plot(data_dict['FHL_datestr'][mask6], data_dict['FHL_T'][mask6], label='FHL_T')

plt.title(f'Temperature Readings from Multiple Sensors\n({start_date.date()} to {end_date.date()})')
plt.xlabel('Date Time')
plt.ylabel('Temperature (°C)')

plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
